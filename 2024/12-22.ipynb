```python
def next_secret(x):
    x = (x*64 ^ x) % 16777216
    x = (x//32 ^ x) % 16777216
    x = (x*2048 ^ x) % 16777216 
    return x
```

```python
def iterate(secret, multiple=2000):
    for i in range(multiple):
        secret = next_secret(secret)
    return secret
```
