```python
from itertools import product
```

```python
with open('data/12-25.input') as f:
    data = f.read()
```

```python
patterns = data.split('\n\n')
```

## Part I

```python
keys = []
locks = []

for p in patterns: 
    rows = p.split()
    if rows[0] == '#####' and rows[-1] == '.....':  #  found a lock
        heights = [col.count('#') - 1 for col in zip(*rows)]
        locks.append(tuple(heights))
    if rows[-1] == '#####' and rows[0] == '.....':  #  found a key
        heights = [col.count('#') - 1 for col in zip(*rows)]
        keys.append(tuple(heights))
```

```python
def fit(key, lock):
    return all(k + l <= 5 for k, l in zip(key, lock))

possible_fit = sum(fit(key, lock) for key in keys for lock in locks)
```
