```python
with open('data/03-Dec.input', 'r') as f:
    data = f.read().strip()
```

```python
data = [[int(y) for y in x.strip()] for x in data.split('\n')]
```

## Part 1

```python
total = 0

for line in data:
    max_so_far = 0
    for i, item in enumerate(line[:-1]):
        max_to_right = max(x for x in line[i+1:])
        max_so_far = max(max_so_far, item*10 + max_to_right)
    total = total + max_so_far

print(total)
```

## Part 2

A different approach... but based on the same idea.  Find the left-most largest digit possible (keeping in mind the total number of digits needed in the resulting number.

```python
def max_number_possible(line, digits=1):
    if digits == 1: 
        return max(line)
    left_most = max(line[:-(digits-1)])
    left_pos = line.index(left_most)
    return left_most * 10**(digits-1) + max_number_possible(line[left_pos+1:], digits-1)
```

Does this still work for Part 1?

```python
total = 0

for d in data:
    total = total + max_number_possible(d, digits=2)
    
print(total)
```

Now do it for 12-digit numbers.

```python
total = 0

for d in data:
    total = total + max_number_possible(d, digits=12)
    
print(total)
```
