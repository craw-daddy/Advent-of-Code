```python
from itertools import combinations
```

```python
with open('data/05-Dec.input', 'r') as f:
    data = f.read().strip()
```

```python
data = data.split('\n')

fresh = []
available = []

for line in data:
    if line:
        if '-' in line:
            fresh.append(tuple([int(x) for x in line.split('-')]))
        else:
            available.append(int(line))
```

```python
fresh[:10]
```

## Part 1

```python
fresh_and_available = sum(any(f[0] <= i <= f[1] for f in fresh) for i in available)

print(fresh_and_available)
```

## Part 2

```python
finished = False

while not finished:
    found = False
    for i,j in combinations(fresh, 2):
        if (i[1] < j[0]) or (j[1] < i[0]):
            pass
        else:
            found = True
            break
    if not found:
        finished = True
    else:
        new_interval = (min(i[0], j[0]), max(i[1], j[1]))
        fresh = [new_interval] + [x for x in fresh if x not in [i,j]]

total = 0
for f in fresh:
    total = total + f[1] - f[0] + 1

print(total)
```
