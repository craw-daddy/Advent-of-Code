```python
with open('data/01-Dec.input', 'r') as f:
    data = f.read()

data = [x.strip() for x in data.split('\n') if x.strip()]
```

# Part 1

```python
hit_zero = 0

dial = 50

for move in data:
    direction, amount = move[0], int(move[1:])
    if direction == 'L':
        while amount > 0:
            if dial == 0:
                dial = 99
            else:
                dial = dial - 1
            amount = amount - 1
    else:
        while amount > 0:
            if dial == 99:
                dial = 0
            else:
                dial = dial + 1
            amount = amount - 1
    if dial == 0:
        hit_zero = hit_zero + 1

print(hit_zero)
```
