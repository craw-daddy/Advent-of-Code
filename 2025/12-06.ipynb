```python
import re

import numpy as np

from functools import reduce
```

```python
with open('data/06-Dec.input', 'r') as f:
    data = f.read().strip()

data = [x.strip() for x in data.split('\n')]
```

## Part 1

(Having read the instructions for Part 2), the data will have to be re-processed since we are basically ignoring spaces here.)

```python
#  Extract numbers as columns
numbers = [[int(x) for x in re.split(r'\s+', line)] for line in data[:-1]]

#  Convert to numpy array and transpose
numbers = np.array(numbers).T

#  Then convert back to list of lists, so to use zip to compute the results
numbers = [[int(x) for x in row] for row in numbers]
```

```python
#  Extract the operator for each set of numbers
operators = [x for x in re.split(r'\s+', data[-1])]
```

```python
results = []

for n, op in zip(numbers, operators):
    if op == '+':
        results.append(reduce(lambda x, y: x+y, n))
    else:
        results.append(reduce(lambda x, y: x*y, n))
```

```python
reduce(lambda x, y: x+y, results)
```

## Part 2

As stated above, we have to re-process the data so as to extract the numbers appropriately.

```python
with open('data/06-Dec.input', 'r') as f:
    data = f.read().strip()

data = [x.strip() for x in data.split('\n')]
```

```python
#  Extract numbers as columns
rows = data[:-1]

#  Extract the operator for each set of numbers
operators = [x for x in re.split(r'\s+', data[-1])]
```

```python
values = []

for k in zip(*rows):
    values.append(''.join(k).strip())
```

```python
size_collection = len(rows)

total = 0
collection = 0
while values:
    nums = []
    while values and values[0]:
        nums.append(int(values[0]))
        values = values[1:]
    if operators[collection] == '+':
        total = total + reduce(lambda x,y: x+y, nums)
    else:
        total = total + reduce(lambda x,y: x*y, nums)
    collection = collection + 1
    values = values[1:]
    
print(total)
```
