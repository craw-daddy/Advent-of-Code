```python
import numpy as np

from itertools import product
```

```python
with open('data/04-Dec.input', 'r') as f:
    data = f.read().strip()

data = np.array([list(x.strip()) for x in data.split('\n')])
```

```python
directions = [np.array(x) for x in [[1,0], [0,1], [0,-1], [-1,0], [1,1], [1,-1], [-1,1], [-1,-1]]]
```

## Part 1

```python
num_neighbors = []

for pos in product(range(data.shape[0]), range(data.shape[1])):
    if data[*pos] == '@':  #  If there is a roll of paper at a location
                           #  then check if it has "few" neighbors
        test = []
        for d in directions:
            v = pos + d
            if  (0 <= v[0] < data.shape[0]) and (0 <= v[1] < data.shape[1]):
                test.append(v)
        num_neighbors.append(sum(data[*v] == '@' for v in test))

print(sum(x < 4 for x in num_neighbors))
```

## Part 2

```python
total_removed = 0

remove = []

#  Find the first set to remove
for pos in product(range(data.shape[0]), range(data.shape[1])):
    if data[*pos] == '@':  #  If there is a roll of paper at a location
                           #  then check if it has "few" neighbors
        test = []
        for d in directions:
            v = pos + d
            if  (0 <= v[0] < data.shape[0]) and (0 <= v[1] < data.shape[1]):
                test.append(v)
        if sum(data[*v] == '@' for v in test) < 4:
            remove.append(pos)

#  Then, while there is stuff to remove, remove it, and check what remains
while len(remove) > 0:
    total_removed = total_removed + len(remove)
    for v in remove:
        data[*v] = 'x'
    remove = []
    
    for pos in product(range(data.shape[0]), range(data.shape[1])):
        if data[*pos] == '@':  #  If there is a roll of paper at a location
                               #  then check if it has "few" neighbors
            test = []
            for d in directions:
                v = pos + d
                if  (0 <= v[0] < data.shape[0]) and (0 <= v[1] < data.shape[1]):
                    test.append(v)
            if sum(data[*v] == '@' for v in test) < 4:
                remove.append(pos)

print(total_removed)
```
