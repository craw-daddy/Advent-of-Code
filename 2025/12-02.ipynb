```python
import re
```

```python
with open('data/02-Dec.input', 'r') as f:
    data = f.read().strip()
```

```python
data = [(int(x.split('-')[0]), int(x.split('-')[1])) for x in data.split(',')]
```

#  Part 1

```python
total = 0

for d in data:
    for i in range(d[0], d[1]+1):
        m = re.search(r'^(\d+)\1$', str(i))
        if m:
            total = total + i

print(total)
```

# Part 2

```python
total = 0

for d in data:
    for i in range(d[0], d[1]+1):
        m = re.search(r'^(\d+)\1+$', str(i))
        if m:
            total = total + i

print(total)
```
