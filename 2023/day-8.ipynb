```python
with open('input/day-8.input', 'r') as f:
    raw = [s.strip() for s in f.readlines()]
```

```python
leftright = raw[0]
mappings = raw[2:]
```

Create a dictionary to lookup a string, and get a tuple of possible locations to move to in the next step.

```python
pairs = {}
for s in mappings:
    partial = s.split(' = ')
    pairs[partial[0]] = tuple([partial[1].split(', ')[0][1:], partial[1].split(', ')[1][:-1]])
```

Map the 'L' and 'R' to 0 and 1 to use as indices to look at tuples.  

```python
leftright = [int(x) for x in list(leftright.replace('L', '0').replace('R', '1'))]
```

## Part I

```python
def walk(start):
    current = start
    ind = 0
    steps = 0

    while current != 'ZZZ':
        current = pairs[current][leftright[ind]]
        ind += 1
        steps += 1
        if ind == len(leftright):
            ind = 0
        
    return steps
```

```python
walk('AAA')
```

## Part II

```python
start = [s for s in pairs.keys() if s[-1] == 'A']
```

```python
def walk_multiple(start):
    current = start.copy()
    ind = 0
    steps = 0
    
    while not all(s[-1] == 'Z' for s in current):
        current = [pairs[c][leftright[ind]] for c in current]
        ind += 1
        steps += 1
        if steps % 10000000 == 0:
            print(current)
        if ind == len(leftright):
            ind = 0
            
    return steps
```
