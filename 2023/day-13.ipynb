## Part I

This is basically just doing what is asked, keeping in mind that we want the number of all rows to the left of the "reflection" column or above the "reflection" row, even if those rows have no corresponding matching column/row in the reflection.  

```python
def find_col(arr):
    potential = []

    for col in range(arr.shape[1]-1):
        if np.all(arr[:,col] == arr[:,col+1]):
            potential.append(col)
            
    if len(potential) == 0:
        return 0
    
    for p in sorted(potential, reverse=True):
        if p == (arr.shape[1] - 1):
            return arr.shape[1] - 1
        if p == 1:
            return 1
    
        test = min(p, arr.shape[1] - p - 2)
    
        if np.all([np.all(arr[:,p-i] == arr[:,p+1+i]) for i in range(1, test+1)]):
            return p + 1
    
    return 0
```

```python
def find_row(arr):
    potential = []

    for row in range(arr.shape[0]-1):
        if np.all(arr[row,:] == arr[row+1,:]):
            potential.append(row)

    if len(potential) == 0:
        return 0
    
    for p in sorted(potential, reverse=True):
        if potential == (arr.shape[0] - 1):
            return arr.shape[0] - 1
        if p == 1:
            return 1
        
        test = min(p, arr.shape[0] - p - 2)
    
        if np.all([np.all(arr[p-i,:] == arr[p+1+i,:]) for i in range(1, test+1)]):
            return p + 1
    
    return 0
```

```python
def signature(arr):
    return 100*find_row(arr) + find_col(arr)
```

```python
total = 0

for a in arrays:
    total += signature(a)
    
total
```
